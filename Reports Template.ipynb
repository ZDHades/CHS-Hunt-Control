```python
import pandas as pd
```

```python
df = pd.read_excel('.xlsx')
df = df.rename(columns=lambda x:x.replace('(Hunt)', 'H'))
df = df.rename(columns=lambda x:x.replace('(Purchase)', 'P'))
df = df.rename(columns=lambda x:x.replace(' ', '_'))
report = df[['Name', 'Level_1_H', 'Level_2_H', 'Level_3_H', 'Level_4_H', 'Level_5_H', 'Level_1_P', 'Level_2_P', 'Level_3_P', 'Level_4_P', 'Level_5_P']]
```

```python
class Guild():
    roster = []
    naughty_list = []
    def __init__(self, name, hunt_1, hunt_2, hunt_3, hunt_4, hunt_5, purchase_white, purchase_green, purchase_blue, purchase_purple, purchase_gold):
        self.name = name
        self.hunt_1 = hunt_1
        self.hunt_2 = hunt_2
        self.hunt_3 = hunt_3
        self.hunt_4 = hunt_4
        self.hunt_5 = hunt_5
        self.purchase_white = purchase_white
        self.purchase_green = purchase_green
        self.purchase_blue = purchase_blue
        self.purchase_purple = purchase_purple
        self.purchase_gold = purchase_gold
        self.hunt = self.get_total_member_hunts()
        self.purchased = self.get_total_member_purchases()
        self.total = self.get_member_total_gifts()
        self.score = self.get_member_score()
        self.roster.append(self)
        
    def __repr__(self):
        return self.name
        
    def am_i_naughty(self):
        if self.score < 70:
            self.naughty_list.append(self)  
    
    def get_total_member_hunts(self):
        return (self.hunt_2 + self.hunt_3 + self.hunt_4 + self.hunt_5)
    
    def get_total_member_purchases(self):
        return (self.purchase_white + self.purchase_green + self.purchase_blue + self.purchase_purple + self.purchase_gold)
    
    def get_member_total_gifts(self):
        return (self.hunt + self.purchased)
    
    def get_member_score(self):
        return self.hunt_2*2 + self.hunt_3*5 + self.hunt_4*7 + self.hunt_5 *10 + self.purchase_white + self.purchase_green*2 + self.purchase_blue*3 + self.purchase_purple*4 +self.purchase_gold*5
    
    @classmethod
    def find_member(self, member):
        for _ in self.roster:
            if member == _.name:
                return _
    
    @classmethod
    def top_hunter(self):
        hunt = 0
        for _ in self.roster:
            if _.name != 'Total' and _.name != 'Anonymous' and _.hunt > hunt: hunt = _.hunt
        for _ in self.roster:
            if _.hunt == hunt:
                return _
            
    @classmethod
    def get_top_hunter(self):
        _ = self.top_hunter()
        print(f"""Top Hunter Breakdown
        Name: {_}
        Total Gifts: {_.total}
        Hunts: {_.hunt}
        Level 1s: {_.hunt_1}
        Level 2s: {_.hunt_2}
        Level 3s: {_.hunt_3}
        Level 4s: {_.hunt_4}
        Level 5s: {_.hunt_5}
        Total Points: {round(_.score, 2)}""")
        self.space_decorator()
    
    @classmethod
    def top_buyer(self):
        purchases = 0
        for _ in self.roster:
            if _.name != 'Total' and _.name != 'Anonymous' and _.purchased > purchases: purchases = _.purchased
        for _ in self.roster:
            if _.purchased == purchases:
                return _
            
    @classmethod
    def get_top_buyer(self):
        _ = self.top_buyer()
        print(f"""Top Buyer Breakdown
        Name: {_}
        Total Gifts: {_.total}
        Packs: {_.purchased}
        White: {_.purchase_white}
        Green: {_.purchase_green}
        Blue: {_.purchase_blue}
        Purple: {_.purchase_purple}
        Gold: {_.purchase_gold}
        Total Points: {round(_.score, 2)}""")
        self.space_decorator()
        
    @classmethod
    def top_gifts(self):
        gifts = 0
        for _ in self.roster:
            if _.name != 'Total' and _.name != 'Anonymous' and _.total > gifts: gifts = _.total
        for _ in self.roster:
            if _.total == gifts: 
                return _
    @classmethod
    def get_top_gifts(self):
        _ = self.top_gifts()
        print(f"""Top Gifter Breakdown
        Name: {_}
        Hunts: {_.hunt}
        Packs: {_.purchased}
        Total Gifts: {_.total}
        Total Points: {round(_.score, 2)}""")
        self.space_decorator()
            
    @classmethod
    def top_points(self):
        points = 0
        for _ in self.roster:
            if _.name != 'Total' and _.name != 'Anonymous' and _.score > points: points = _.score
        for _ in self.roster:
            if _.score == points:
                return _
            
    @classmethod
    def get_top_points(self):
        _ = self.top_points()
        print(f"""Top Points Breakdown
        Name: {_}
        Hunts: {_.hunt}
        Packs: {_.purchased}
        Total Gifts: {_.total}
        Total Points: {round(_.score, 2)}""")
        self.space_decorator()
    
    @classmethod 
    def overall_breakdown(self, _):
        print(f"""{_} Breakdown
        Gifts: {_.total}
        Hunts: {_.hunt}
        Purchases: {_.purchased}
        Level 1s: {_.hunt_1}
        Level 2s: {_.hunt_2}
        Level 3s: {_.hunt_3}
        Level 4s: {_.hunt_4}
        Level 5s: {_.hunt_5}
        White: {_.purchase_white}
        Green: {_.purchase_green}
        Blue: {_.purchase_blue}
        Purple: {_.purchase_purple}
        Gold: {_.purchase_gold}
        Total Points: {_.score}""")
    
    @classmethod
    def average_breakdown(self, _):
        print(f"""Daily average Breakdown
    Total Gifts: {round(_.total/7, 2)}
    Hunts: {round(_.hunt/7, 2)}
    Purchases: {round(_.purchased/7, 2)}
    Level 1s: {round(_.hunt_1/7, 2)}
    Level 2s: {round(_.hunt_2/7, 2)}
    Level 3s: {round(_.hunt_3/7, 2)}
    Level 4s: {round(_.hunt_4/7, 2)}
    Level 5s: {round(_.hunt_5/7, 2)}
    Total Points: {round(_.score/7, 2)}""")
        
    @classmethod
    def get_total_gifts(self):
        self.overall_breakdown(self.find_member('Total'))
        self.space_decorator()
        
    @classmethod
    def get_anon_gifts(self):
        self.overall_breakdown(self.find_member('Anonymous'))
        self.space_decorator()
    
    @classmethod
    def get_average(self):
        self.average_breakdown(self.find_member('Total'))
        self.space_decorator()
        
    @classmethod
    def get_naughty_list(self):
        print("""-What it means to be on the naughty list-
    For each hunt report, we translate the gifts from each member into points:
    Loot                |  Bonus 
    level 1 : 0 points  |  White  : 1 point
    Level 2 : 2 points  |  Green  : 2 points
    Level 3 : 5 points  |  Blue   : 3 points
    Level 4 : 7 points  |  Purple : 4 points
    Level 5 : 10 points |  Gold   : 5 points
    10 points per day required - 70 points per week.
    Each week starts at midnight EST on Sunday.
    If your name is listed below, it means that you did not reach the gift requirements. 
    The number shown is the amount of points you are missing for this week.""")
        print()
        print(f"There are {len(self.naughty_list)} members on the naughty list")
        print()
        for _ in self.naughty_list:
            print(f"{_} is missing {70-_.score} points worth of gifts for this week.")
    
    @classmethod
    def space_decorator(self):
        print()
        print("="*40)
        print()
```
