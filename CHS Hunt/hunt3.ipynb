```python
df1 = pd.read_excel(xls, '06')
df2 = pd.read_excel(xls, '07')
df3 = pd.read_excel(xls, '08')
df4 = pd.read_excel(xls, '09')
df5 = pd.read_excel(xls, '10')
df6 = pd.read_excel(xls, '11')

frame = [df1, df2, df3, df4, df5, df6]
Hunt = pd.concat(frame)
Hunt = Hunt.reset_index(drop=True)
Hunt = Hunt.rename(columns=lambda x:x.replace(' ', '_'))
```

```python
class Guild():
    roster = []
    def __init__(self, name, total, hunt, purchased, one, two, three, four, five):
        self.name = name
        self.total = total
        self.hunt = hunt
        self.purchased = purchased
        self.one = one
        self.two = two
        self.three = three
        self.four = four
        self.five = five
        self.score = self.get_score()
        self.roster.append(self)
        
    def __repr__(self):
        return self.name       
    
    def get_score(self):
        return ((self.purchased * .5) + (self.two * .2) + (self.three * .5) + self.four + self.five)
```

```python
# Initalize player instances
def get_roster():
    roster = []
    for _ in Hunt.Name:
        if _ not in roster:
            roster.append(_)
    for _ in roster:
        total = Hunt.query(f'Name == "{_}"').Total.sum()
        hunt = Hunt.query(f'Name == "{_}"').Hunt.sum()
        purchased = Hunt.query(f'Name == "{_}"').Purchase.sum()
        one = Hunt.query(f'Name == "{_}"').Level_1.sum()
        two = Hunt.query(f'Name == "{_}"').Level_2.sum()
        three = Hunt.query(f'Name == "{_}"').Level_3.sum()
        four = Hunt.query(f'Name == "{_}"').Level_4.sum()
        five = Hunt.query(f'Name == "{_}"').Level_5.sum()
        g = Guild(name=_, total=total, hunt=hunt, purchased=purchased, one=one, two=two, three=three, four=four, five=five)
```

```python
# Finding the top Hunter
def top_hunter(roster):
    hunts = 0
    for _ in roster:
        if _.name != 'Total' and _.name != 'Anonymous' and _.hunt > hunts: hunts = _.hunt
    for _ in roster:
        if _.hunt == hunts:
            return _
        
# Find the top Buyer
def top_buyer(roster):
    purchases = 0
    for _ in roster:
        if _.name != 'Total' and _.name != 'Anonymous' and _.purchased > purchases: purchases = _.purchased
    for _ in roster:
        if _.purchased == purchases:
            return _
        
# Find the top Gifter
def top_gifter(roster):
    gifts = 0
    for _ in roster:
        if _.name != 'Total' and _.name != 'Anonymous' and _.total > gifts: gifts = _.total
    for _ in roster:
        if _.total == gifts: 
            return _
```

```python
# Breakdowns for categories
def breakdown(_, message):
    print(f"""{message} breakdown
    {message}: {_}
    Total Gifts: {_.total}
    Hunts: {_.hunt}
    Purchases: {_.purchased}
    Level 1s: {_.one}
    Level 2s: {_.two}
    Level 3s: {_.three}
    Level 4s: {_.four}
    Level 5s: {_.five}""")

# overall Breakdowns
def overall_breakdown(_):
    print(f"""{_} breakdown
    Total Gifts: {_.total}
    Hunts: {_.hunt}
    Purchases: {_.purchased}
    Level 1s: {_.one}
    Level 2s: {_.two}
    Level 3s: {_.three}
    Level 4s: {_.four}
    Level 5s: {_.five}""")
```

```python
# Find a player
def find_player(player, roster):
    for _ in roster:
        if player == _.name:
            return _
```

```python
# Finding the Naughty List
class Naughty():
    naughty_list = []
    def __init__(self, name, total, score, hunt, purchased, one, two, three, four, five):
        self.name = name
        self.total = total
        self.hunt = hunt
        self.purchased = purchased
        self.one = one
        self.two = two
        self.three = three
        self.four = four
        self.five = five
        self.score = score
        self.naughty_list.append(self)
    
    def __repr__(self):
        return f"{self.name}"
    
    @classmethod
    def get_naughty_list(self, days):
        print(f"There are {len(self.naughty_list)} players on the naughty list")
        print()
        for _ in self.naughty_list:
            print(f"{_} is missing {round(days-_.score, 2)} days worth of completed hunts.")
        
def make_naughty_list(roster, days):
    for _ in roster:
        if _.score < days:
            Naughty(name=_.name, score=_.score, total=_.total, hunt=_.hunt, purchased=_.purchased, one=_.one, two=_.two, three=_.three, four=_.four, five=_.five)
```

```python
def space_decorater():
    print()
    print("="*40)
    print()
```

```python
def get_report(days):
    get_roster()
    make_naughty_list(Guild.roster, days)
    breakdown(top_hunter(Guild.roster), "Top Hunter")
    space_decorater()
    breakdown(top_buyer(Guild.roster), "Top Buyer")
    space_decorater()
    breakdown(top_gifter(Guild.roster), "Top Gifter")
    space_decorater()
    overall_breakdown(find_player("Total", Guild.roster))
    space_decorater()
    overall_breakdown(find_player("Anonymous", Guild.roster))
    space_decorater()
    Naughty.get_naughty_list(days)
```
