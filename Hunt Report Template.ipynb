```python
class Guild():
    roster = []
    naughty_list = []
    def __init__(self, name, h1, h2, h3, h4, h5, p1, p2, p3, p4, p5):
        self.name = name
        self.h1 = h1
        self.h2 = h2
        self.h3 = h3
        self.h4 = h4
        self.h5 = h5
        self.p1 = p1
        self.p2 = p2
        self.p3 = p3
        self.p4 = p4
        self.p5 = p5
        self.hunt = self.get_hunt()
        self.purchased = self.get_purchased()
        self.total = self.get_total()
        self.score = self.get_score()
        self.roster.append(self)
        
    def am_i_naughty(self):
        if self.score < 70:
            self.naughty_list.append(self)
        
    def __repr__(self):
        return self.name       
    
    def get_hunt(self):
        return (self.h2 + self.h3 + self.h4 + self.h5)
    
    def get_purchased(self):
        return (self.p1 + self.p2 + self.p3 + self.p4 + self.p5)
    
    def get_total(self):
        return (self.hunt + self.purchased)
    
    def get_score(self):
        return self.h2*2 + self.h3*5 + self.h4*7 + self.h5 *10 + self.p1 + self.p2*2 + self.p3*3 + self.p4*4 +self.p5*5
    
    @classmethod
    def get_naughty_list(self):
        print("""-What it means to be on the naughty list-
    For each hunt report, we translate the gifts from each player into points:
    Loot                |  Bonus 
    level 1 : 0 points  |  White  : 1 point
    Level 2 : 2 points  |  Green  : 2 points
    Level 3 : 5 points  |  Blue   : 3 points
    Level 4 : 7 points  |  Purple : 4 points
    Level 5 : 10 points |  Gold   : 5 points
    10 points per day required - 70 points per week.
    Each week starts at midnight EST on Sunday.
    If your name is listed below, it means that you did not reach the gift requirements. 
    The number shown is the amount of points you are missing for this week.""")
        print()
        print(f"There are {len(self.naughty_list)} players on the naughty list")
        print()
        for _ in self.naughty_list:
            print(f"{_} is missing {70-_.score} points worth of gifts for this week.")
```

```python
def get_roster():
    roster = []
    for _ in report.Name:
        roster.append(_)
    for _ in roster:
        h1 = report.query(f'Name == "{_}"').Level_1_H.sum()
        h2 = report.query(f'Name == "{_}"').Level_2_H.sum()
        h3 = report.query(f'Name == "{_}"').Level_3_H.sum()
        h4 = report.query(f'Name == "{_}"').Level_4_H.sum()
        h5 = report.query(f'Name == "{_}"').Level_5_H.sum()
        p1 = report.query(f'Name == "{_}"').Level_1_P.sum()
        p2 = report.query(f'Name == "{_}"').Level_2_P.sum()
        p3 = report.query(f'Name == "{_}"').Level_3_P.sum()
        p4 = report.query(f'Name == "{_}"').Level_4_P.sum()
        p5 = report.query(f'Name == "{_}"').Level_5_P.sum()
        g = Guild(name=_, h1=h1, h2=h2, h3=h3, h4=h4, h5=h5, p1=p1, p2=p2, p3=p3, p4=p4, p5=p5)
```

```python
# Finding the top Hunter
def top_hunter(roster):
    hunt = 0
    for _ in roster:
        if _.name != 'Total' and _.name != 'Anonymous' and _.hunt > hunt: hunt = _.hunt
    for _ in roster:
        if _.hunt == hunt:
            return _
        
# Find the top Buyer
def top_buyer(roster):
    purchases = 0
    for _ in roster:
        if _.name != 'Total' and _.name != 'Anonymous' and _.purchased > purchases: purchases = _.purchased
    for _ in roster:
        if _.purchased == purchases:
            return _
        
# Find the top Gifter
def top_gifter(roster):
    gifts = 0
    for _ in roster:
        if _.name != 'Total' and _.name != 'Anonymous' and _.total > gifts: gifts = _.total
    for _ in roster:
        if _.total == gifts: 
            return _
# Find the player with the most points
def top_points(roster):
    points = 0
    for _ in roster:
        if _.name != 'Total' and _.name != 'Anonymous' and _.score > points: points = _.score
    for _ in roster:
        if _.score == points:
            return _
```

```python
# Breakdowns for categories
def top_hunter_breakdown(_):
    print(f"""Top Hunter Breakdown
    Name: {_}
    Total Gifts: {_.total}
    Hunts: {_.hunt}
    Level 1s: {_.h1}
    Level 2s: {_.h2}
    Level 3s: {_.h3}
    Level 4s: {_.h4}
    Level 5s: {_.h5}
    Total Points: {round(_.score, 2)}""")
    
def top_packs_breakdown(_):
    print(f"""Top Buyer Breakdown
    Name: {_}
    Total Gifts: {_.total}
    Packs: {_.purchased}
    White: {_.p1}
    Green: {_.p2}
    Blue: {_.p3}
    Purple: {_.p4}
    Gold: {_.p5}
    Total Points: {round(_.score, 2)}""")

def top_gifter_breakdown(_):
    print(f"""Top Gifter Breakdown
    Name: {_}
    Hunts: {_.hunt}
    Packs: {_.purchased}
    Total Gifts: {_.total}
    Total Points: {round(_.score, 2)}""")
    
def top_points_breakdown(_):
    print(f"""Top Points Breakdown
    Name: {_}
    Hunts: {_.hunt}
    Packs: {_.purchased}
    Total Gifts: {_.total}
    Total Points: {round(_.score, 2)}""")

# overall Breakdown
def overall_breakdown(_):
    print(f"""{_} Breakdown
    Gifts: {_.total}
    Hunts: {_.hunt}
    Purchases: {_.purchased}
    Level 1s: {_.h1}
    Level 2s: {_.h2}
    Level 3s: {_.h3}
    Level 4s: {_.h4}
    Level 5s: {_.h5}
    White: {_.p1}
    Green: {_.p2}
    Blue: {_.p3}
    Purple: {_.p4}
    Gold: {_.p5}
    Total Points: {round(_.score, 2)}""")

# session average Breakdown
def session_average(_):
    print(f"""7 day report average Breakdown
    Total Gifts: {round(_.total/7, 2)}
    Hunts: {round(_.hunt/7, 2)}
    Purchases: {round(_.purchased/7, 2)}
    Level 1s: {round(_.h1/7, 2)}
    Level 2s: {round(_.h2/7, 2)}
    Level 3s: {round(_.h3/7, 2)}
    Level 4s: {round(_.h4/7, 2)}
    Level 5s: {round(_.h5/7, 2)}
    Total Points: {round(_.score, 2)}""")
```

```python
# Find a player
def find_player(player, roster):
    for _ in roster:
        if player == _.name:
            return _

def space_decorator():
    print()
    print("="*40)
    print()
```

```python
def get_report():
    get_roster()
    for _ in Guild.roster:
        _.am_i_naughty() 
    top_hunter_breakdown(top_hunter(Guild.roster))
    space_decorator()
    top_packs_breakdown(top_buyer(Guild.roster))
    space_decorator()
    top_gifter_breakdown(top_gifter(Guild.roster))
    space_decorator()
    top_points_breakdown(top_points(Guild.roster))
    space_decorator()
    overall_breakdown(find_player("Anonymous", Guild.roster))
    space_decorator()
    session_average(find_player("Total", Guild.roster))
    space_decorator()
    Guild.get_naughty_list()
```
